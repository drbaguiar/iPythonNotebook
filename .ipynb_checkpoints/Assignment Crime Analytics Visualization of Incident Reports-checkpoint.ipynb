## Data Preparation
First I'll prepare the workspace with the following commands:

```r
rm(list=ls())

##Turn off scientific notations for numbers
options(scipen = 999)  

##Set locale
Sys.setlocale("LC_ALL", "English") 

##Set seed for reproducibility
set.seed(1234)

# Turn off warnings
options(warn = -1)

# Function clean the data names 
# Use: df<-cleanit(df)
cleanit <-function(df){
  names(df) <-tolower(names(df))
  names(df) <- gsub("\\(","",names(df))
  names(df) <- gsub("\\)","",names(df))
  names(df) <- gsub("\\.","",names(df))
  names(df) <- gsub("_","",names(df))
  names(df) <- gsub("-","",names(df))
  names(df) <- gsub(",","",names(df))
  return(df)
}
```

Next, I'll read the data from GitHub

```r
library(RCurl)
x <- getURL("https://raw.githubusercontent.com/uwescience/datasci_course_materials/master/assignment6/seattle_incidents_summer_2014.csv")
df <- read.csv(text = x)
```

I'll take a look at the structure of the data using the str function in R.

```r
df<-cleanit(df)
str(df)
#head(df)
```

##Analysis

Count by month, sort in descending order

```r
sort(table(df$month),decreasing = TRUE)
```

```r
plot(sort(table(df$month),decreasing = TRUE),type="b",lwd=2, xlab="Month",ylab="Count")
text(1,11100,label="July")
text(2,11100,label="June")
text(2.85,10525,label="August")
```

```r
sort(table(df$offensetype),decreasing = TRUE)
df$nbr<-1
```

###View a barplot of the offense types 
####(Top 8 Offenses Only)

```r
x<-sort(table(df$offensetype),decreasing = TRUE)
y<-head(x, n=8)
barplot(y, main="Top 8 Offense Types",col=terrain.colors(8))
```

Next, I'll display the top offenses by month (6, 7, or 8).  First, I'll extract the data for the top offenses

```r
include <-c("THEFT-CARPROWL","VEH-THEFT-AUTO","THEFT-OTH","PROPERTY DAMAGE-NON RESIDENTIA","ASSLT-NONAGG","DISTURBANCE-OTH","BURGLARY-FORCE-RES","BURGLARY-NOFORCE-RES")
df2<-subset(df, offensetype %in% include)
```

Then I'll separate the data by month.

```r
z <-data.frame(aggregate(nbr ~ month+offensetype, data = df2, sum) )
barplot(z$nbr, col=terrain.colors(3), names.arg=z$offensetype)
legend("topleft", inset=.05, title="Months", c("6","7","8"), fill=terrain.colors(3), horiz=TRUE)
```

##Conclusion

Crime for the top 8 offense types is higher in July in 3 of out of the 8 types. Crime is higher in June for 3 offense types out the 8 top types. Crime is higher in August for 2 offense types out the 8 top types.
