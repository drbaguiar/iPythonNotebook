# Probability Distributions
Probability distribution describes how the values of a random variable is distributed. For example, the collection of all possible outcomes of a sequence of coin tossing is known to follow the binomial distribution. Because the characteristics of these theoretical distributions are well understood, they can be used to make statistical inferences on the entire data population as a whole.

For every distribution there are four commands. The commands for each distribution are prepended with a letter to indicate the functionality:

“d”	returns the height of the probability density function

“p”	returns the cumulative density function

“q”	returns the inverse cumulative density function (quantiles)

“r”	returns randomly generated numbers

##Discrete Distributions

###Binomial Distribution
The binomial distribution is a <B>discrete probability</b> distribution. It describes the outcome of n independent trials in an experiment. Each trial is assumed to have only two outcomes, either success or failure. 

```r
x <- seq(0,50,by=1)
y <- dbinom(x,50,0.2)
plot(x,y)
y <- dbinom(x,50,0.6)
plot(x,y)
x <- seq(0,100,by=1)
y <- dbinom(x,100,0.6)
plot(x,y)
```

### Examples
Suppose there are twelve multiple choice questions in an English class quiz. Each question has five possible answers, and only one of them is correct. Find the probability of having four or less correct answers if a student attempts to answer every question at random.

Because only one out of five possible answers is correct, the probability of answering a question correctly by random is 1/5=0.2. We can find the probability of having <b>exactly 4</B> correct answers by random attempts as follows.

```r
dbinom(4, size=12, prob=0.2) 
```

To find the probability of having four or less correct answers by random attempts, we apply the function dbinom as follows:

```r
dbinom(4, size=12, prob=0.2)+
+dbinom(3, size=12, prob=0.2)+
+dbinom(2, size=12, prob=0.2)+
+dbinom(1, size=12, prob=0.2)+
+dbinom(0, size=12, prob=0.2) 
```

```r
sum(dbinom(0:4, size=12, prob=0.2))
```

Alternatively, we can use the <B><I>cumulative probability</I></B> function for binomial distribution pbinom.

```r
pbinom(4, size=12, prob=0.2) 
```

Compute P(45 < X < 55) for X Binomial(100,0.5)

```r
sum(dbinom(46:54, 100, 0.5))
```

```r
pbinom(54, size=100, prob=0.5) - pbinom(45, size=100, prob=0.5) 
```

##Continous Distributions

###Normal Distribution 

```r
x <- seq(-20,20,by=.1)
y <- dnorm(x)
plot(x,y)
y <- dnorm(x,mean=2.5,sd=0.1)
plot(x,y)
```

### Examples
Assume that the test scores of a college entrance exam fits a normal distribution. Furthermore, the mean test score is 72, and the standard deviation is 15.2. What is the percentage of students scoring 84 or more in the exam?

We apply the function pnorm of the normal distribution with mean 72 and standard deviation 15.2. Since we are looking for the percentage of students scoring higher than 84, we are interested in the upper tail of the normal distribution.

```r
pnorm(84, mean=72, sd=15.2, lower.tail=FALSE)
```
