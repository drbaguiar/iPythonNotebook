### Statistical stationarity
A stationary time series is one whose statistical properties such as mean, variance, autocorrelation, etc. are all constant over time. Most statistical forecasting methods are based on the assumption that the time series can be rendered approximately stationary (i.e., "stationarized") through the use of mathematical transformations. A stationarized series is relatively easy to predict: you simply predict that its statistical properties will be the same in the future as they have been in the past!   When you detect non statioinary, you need to transform or difference the data.  

```r
### Stationarity
library(tseries)
## Null Hypothesis Non Stationary
## Alternative Hypothesis: Stationary
```

```r
x <- rnorm(1000) # no unit-root, stationary
z <- c(1:1000)
adf.test(x) # augmented Dickey Fuller Test
adf.test(z)
```

```r
plot(nottem) # Let s see the nottem dataset
plot(decompose(nottem))
adf.test(nottem)
```

```r
y <- diffinv(x) # non-stationary
plot(y)
adf.test(y)
```

```r
### Autocorrelation (Do previous observation influnece the recent)
##Lags - stepson a time scale
# Durbin Watson test for autocorrelation.  
## Null Hypothesis: No autocorrelation 
## Alternative Hypothesis: There is autocorrelation
# Only tests 1 time period  Use with caution
library(lmtest)
```

```r

### Simple Exp Smoothing
plot(LakeHuron) # looks like a random walk
HoltWinters(LakeHuron, gamma = F, beta = F) # getting the model
plot(HoltWinters(LakeHuron, gamma = F, beta = F))
x = HoltWinters(LakeHuron, gamma = F, beta = F) # object x to forecast
library(forecast) # activate library for further commands
forecast.HoltWinters(x, h = 5)
plot.forecast(forecast.HoltWinters(x, h = 5)) # forecast plot
y = forecast.HoltWinters(x, h = 5) # new object y
acf(y$residuals) # checking if the model works for the dataset
plot.ts(y$residuals)
Box.test(y$residuals, lag=20, type="Ljung-Box")


### Holt Winters Exp Smoothing
plot((AirPassengers)) # trend, seasonality, additive???
HoltWinters(AirPassengers)
plot(HoltWinters(AirPassengers))
x = HoltWinters((AirPassengers))
forecast.HoltWinters(x, h = 12) # one year
plot.forecast(forecast.HoltWinters(x, h = 12))
y = forecast.HoltWinters(x, h = 12)
acf(y$residuals, lag = 20)
plot.ts(y$residuals)
Box.test(y$residuals, lag=20, type="Ljung-Box") # log or diff may be needed
```
